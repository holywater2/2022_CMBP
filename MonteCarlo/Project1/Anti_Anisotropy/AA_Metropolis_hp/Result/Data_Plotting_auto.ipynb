```python
ls = []
# alpha_list = ["a=0.5","a=1.0","a=1.5","a=2.0","a=2.5","a=3.0"]
Llist = ["6_6","8_8","12_12","16_16","32_32"]

for ll in Llist:
    ls.append([ll , pd.read_csv("./auto/AA_Metropolis_c_"+ll+"_int10_mcs33554432_"+"a"+alpha_file_name_dict["a=1.0"]+"_auto_MM_1.csv", skiprows = [0], header=None)])
# tt.mean(axis='rows'), tt.std(axis='rows')

plt.style.use('seaborn-whitegrid')

tau = []
for dd in ls:
    data = dd[1]
    # cur_tau = ((data.mean(axis='rows')[23]/data.mean(axis='rows')[4])**2-1)/2
    cur_tau = ((data.mean(axis='rows')[25]/data.mean(axis='rows')[4])**2)

    tau.append(cur_tau)
    # plt.errorbar(range(0,23),data.mean(axis='rows')[4:],data.std(axis='rows')[4:],label = dd[0]+" $M=2^{23}$, $\\tau=$",marker='s',markersize=5)
    plt.plot(range(0,25),data.mean(axis='rows')[4:],label = dd[0]+" $M=2^{23}$, $\\tau=$"+ str(int(cur_tau)),marker='s',markersize=5)

# plt.ylim(0,0.004)
plt.ylabel('$delta^{(l)}$')
plt.xlabel('binning level(l)')
plt.text(11,0.00002,'a = 0.5')
plt.legend()
print(tau)
plt.show()

from scipy.optimize import curve_fit
from scipy.special import gammaincc
x = [6,8,12,16,32]
y = tau

func = lambda x, a0, a1 : x**a1 + a0
popt, pcov = curve_fit(func,x,y)

plt.scatter(x,y,marker="o",label='alpha=1.0')
xx = np.linspace(0,110)
yy = func(xx,*popt)
plt.plot(xx,yy,linestyle='--')
plt.xlabel("Lattice size: L*L")
plt.ylabel("Integrated autocorrealtion time")
plt.legend()
plt.scatter(100,func(100,*popt))
plt.text(3,8100,"$tau = L^{"+str(round(popt[1],4))+"}+("+str(int(popt[0]))+")$")
```
