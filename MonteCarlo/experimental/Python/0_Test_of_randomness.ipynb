```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bin = 100
epoch = int(10E6)
l = np.zeros(bin)
for i in range(epoch):
    k = np.random.random()
    l[int(k*100)] += 1
```

```python
x = [i/bin for i in range(bin)]

plt.plot(x,l)
plt.show()
```

```python
# Barkema Ch 1.4
def f(x):
    return np.sin(1/x)**2

bin2 = 10000

x2 = [(i+1)/bin2 for i in range(2*bin2)]
y2 = [f((j+1)/bin2) for j in range(2*bin2)]

plt.plot(x2,y2)
plt.show()
```

```python
## width (0 to X) heigh (0 to 1) 크기의 박스를 그리고 (h,v)를 그 안에서 랜덤으로 선택한다.
## 그 다음 (h,v)가 (h, f(h))보다 작을 확룰을 계산하고, 이 확률에다가 상자의 크기를 곱한다.

epoch3 = 1000
bin3 = 100
x3 = [2*(i+1)/bin3 for i in range(bin3)]
y3 = np.zeros(100)
for k in range(bin3):
    X = 2*(k+1)/bin3
    for j in range(epoch3):
        h = np.random.random()*X #(0 to X)
        v = np.random.random()   #(0 to 1)
        if(v < f(h)):
            y3[k] += 1
plt.plot(x3,(x3*y3)/epoch3)
plt.show()
```

```python
epoch4 = 10000000

def qudr(x):
    return np.sqrt(1-x**2)

res = 0

for i in range(epoch4):
    h1 = np.random.random()
    v1 = np.random.random()
    if(v1 < qudr(h1)):
        res += 1

print(4*res/epoch4,np.pi)
```
